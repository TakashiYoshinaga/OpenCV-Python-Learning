```python
import cv2
#matplotlibのpyplotの読み込み。(画像の描画に使う)
from matplotlib import pyplot as plt

#画像を読み込む。第二引数は 0:グレースケール、1:カラー画像 -1:カラー＋アルファ
img = cv2.imread('Images/arfukuoka.png',1)
#画像をWindowで表示
plt.imshow(img)
```

```python
#OpenCVはBGRの順番に色が格納されるため色が正しく表示されない。
#そのためBとRの入れ替えを行う。今回は変換結果を再度imgに代入
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
#画像をWindowで表示
plt.imshow(img)
```
